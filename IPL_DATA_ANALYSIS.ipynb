```python
match_data['seasons']=pd.DatetimeIndex(match_data['date']).year
print("seasons played so far :\n",match_data['seasons'].unique())
```

Lets see matches played per season

```python
matches_per_season=match_data.groupby(['seasons'])['id'].count().reset_index().rename(columns={'id':'matches'})
```

```python
matches_per_season
```

Now lets join the season column from matche_data to ball_data so that we can perform some more analysis

```python
season_data=match_data[['id','seasons']].merge(ball_data, left_on='id' , right_on='id', how='left').drop('id',axis=1)
```

```python
season_data.head()
```

Now we"ll see the total runs scored in each season

```python
runs_per_season=season_data.groupby(['seasons'])['total_runs'].count().reset_index()
runs_per_season
```

Now lets see avg of total runs scored in a match in each season


```python
runs_per_match=pd.concat([runs_per_season,matches_per_season],axis=1)
runs_per_match
```

```python
runs_per_match['runs_per_match']=runs_per_season['total_runs']/matches_per_season['matches']
runs_per_match
```

Its time to see the best batsman of Ipl So far

```python
most_runs=season_data.groupby(['batsman'])['batsman_runs'].sum().reset_index()
most_runs.columns=['batsman','runs']
most_runs=most_runs.sort_values(by='runs',ascending = False).head(10).reset_index().drop(columns='index')
most_runs
```

The above is the list of top 10 batsman in the history of IPL and VIRAT KOHLI is the best batsman so far
